```python
from scipy.stats import (uniform, binom, poisson, norm, expon)
import matplotlib.pyplot as plt
import numpy as np
```

# Равномерное распределение

\begin{align*}
& x \sim U(a,b) \\
& \textbf{Плотность}:  f(x)=\begin{cases}
                                \frac{1}{b-a}, & x \in [a,b]\\
                                0, & x \notin [a,b]\\
                            \end{cases} \\
& \textbf{Функция распределения}: F(x) = \begin{cases}
                                             0, & x < a \\
                                             \frac{x-a}{b-a}, & x \in [a,b)\\
                                             1, & x \geq b \\
                                         \end{cases}  \\
& \textbf{Математическое ожидание}: \mathbb{E} = \frac{a+b}{2} \\
& \textbf{Дисперсия}: \mathbb{D} = \frac{(b-a)^2}{12} \\
\end{align*}

Параметры, зададающие равномерное распределение:
- loc -- Математическое ожидание
- scale -- Среднеквадратичное отклонение

```python
x = np.linspace(-2, 20, 2000)
```

```python
for i in range(5):
    plt.plot(x, uniform.pdf(x, 0, 2**i))
plt.title('Плотность равномерного распределения')
plt.legend(['scale = {}'.format(2**i) for i in range(5)])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
for i in range(5):
    plt.plot(x, uniform.cdf(x, 0, 2**i))
plt.title('Функция распределения равномерного распределения')
plt.legend(['scale = {}'.format(2**i) for i in range(5)])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

# Биномиальное распределение

\begin{align*}
& x \sim B(n,p) \\
& q = 1-p \\
& \textbf{Функция вероятности}: P(x = k) = C^k_n p^k q^{n-k} \\
& \textbf{Математическое ожидание}: \mathbb{E} = np \\
& \textbf{Дисперсия}: \mathbb{D} = npq
\end{align*}

Параметры, зададающие биномиальное распределение:
- n -- Число испытаний
- p -- Вероятность успеха

```python
x = np.arange(0, 26)
```

```python
for i in range(3):
    plt.plot(x, binom.pmf(x, 50, (i+1)/10), '-o')
plt.title('Распределение вероятностей дискретной случайной величины биномиального распределения')
plt.legend(['p = {}'.format((i+1)/10) for i in range(3)])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
for i in range(3):
    plt.plot(x, binom.cdf(x, 50, (i+1)/10), '-o')
plt.title('Функция распределения биномиального распределения')
plt.legend(['p = {}'.format((i+1)/10) for i in range(3)])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

# Распределение Пуассона

\begin{align*}
& x \sim P(\lambda), \lambda \in (0, + \infty) \\
& \Gamma(s,x) = \int_x^\infty t^{s-1} e^{-t} dt\\
& \textbf{Функция распределения}: F(k) = \frac{\Gamma(k+1, \lambda)}{k!}, \lambda > 0 \\
& \textbf{Функция вероятности}: P(x = k) = \frac{e^{-\lambda}\lambda^k}{k!}, \lambda > 0 \\
& \textbf{Математическое ожидание}: \mathbb{E} = \lambda \\
& \textbf{Дисперсия}: \mathbb{D} = \lambda
\end{align*}

Параметры, зададающие распределение Пуассона:
- mu -- Математическое ожидание

```python
x = np.arange(-2, 13)
```

```python
for i in range(4):
    plt.plot(x, poisson.pmf(x, mu=i+1), '-o')
plt.title('Распределение вероятностей дискретной случайной величины распределения Пуассона')
plt.legend(['mu = {}'.format(i+1) for i in range(4)])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
for i in range(4):
    plt.plot(x, poisson.cdf(x, mu=i+1), '-o')
plt.title('Функция распределения распределения Пуассона')
plt.legend(['mu = {}'.format(i+1) for i in range(4)])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

# Нормальное распределение

\begin{align*}
& x \sim N(\mu, \sigma^2) \\
& erf(x) = \frac{2}{\pi} \int_0^x e^{-t^2}dt \\
& \textbf{Плотность}: F(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{\frac{-(x-\mu)^2}{2\sigma^2}}\\
& \textbf{Функция распределения}: F(x) = \frac{1}{2} \left(1+erf\left(\frac{x-\mu}{\sqrt{2\sigma^2}}\right)\right) \\
& \textbf{Функция вероятности}: P(x = k) = \frac{e^{-\lambda}\lambda^k}{k!}, \lambda > 0 \\
& \textbf{Математическое ожидание}: \mathbb{E} = \mu \\
& \textbf{Дисперсия}: \mathbb{D} = \sigma^2
\end{align*}

Параметры, зададающие равномерное распределение:
- loc -- Математическое ожидание
- scale -- Среднеквадратичное отклонение

```python
x = np.linspace(-20, 20, 2000)
```

```python
for i in range(5):
    plt.plot(x, norm.pdf(x, 0, 2 ** i))
plt.title('Плотность нормального распределения')
plt.legend(['scale = {}'.format(2 ** i) for i in range(5)])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
for i in range(5):
    plt.plot(x, norm.cdf(x, 0, 2 ** i))
plt.title('Функция распределения нормального распределения')
plt.legend(['scale = {}'.format(2 ** i) for i in range(5)])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

# Экспоненциальное распределение

\begin{align*}
& x \sim Exp(\lambda) \\
& \textbf{Плотность}: f(x) = \begin{cases}
                                 \lambda e^{-\lambda x}, & x \geq 0 \\
                                 0, x<0 \\
                             \end{cases}  \\
& \textbf{Функция распределения}: F(x) = \begin{cases}
                                            1 - e^{-\lambda x}, & x \geq 0 \\
                                            0, x<0 \\
                                  \end{cases}  \\
& \textbf{Математическое ожидание}: \mathbb{E} = \frac{1}{\lambda} \\
& \textbf{Дисперсия}: \mathbb{D} = \frac{1}{\lambda^2}
\end{align*}

Параметры, зададающие экспоненциальное распределение:
- loc -- Сдвиг распределения
- scale -- Масштабирование распределения

```python
x = np.linspace(-2, 15, 2000)
```

```python
for i in range(5):
    plt.plot(x, expon.pdf(x, 0, i+1))
plt.title('Плотность равномерного распределения')
plt.legend(['scale = {}'.format(i+1) for i in range(5)])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
for i in range(5):
    plt.plot(x, expon.cdf(x, 0, i+1))
plt.title('Плотность равномерного распределения')
plt.legend(['scale = {}'.format(i+1) for i in range(5)])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```
